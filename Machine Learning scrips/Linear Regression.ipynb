```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
```

```python
from sklearn import linear_model
```

```python
from keras.layers import Input, Dense
from keras.models import Model
```

```python
path = os.getcwd() + '/data/ex1data1.txt'
data = pd.read_csv(path, names = ['Population','Profit'])
data.head()
```

```python
data.describe()
```

```python
data.plot(kind='Scatter', x = 'Population', y = 'Profit', figsize=(15,10))
```

```python
#Reshape cambia de un array de 97,1 a uno de 1,97 
#populationX = data['Population'].values.reshape(-1,1)
populationX = data['Population'].values.reshape(-1,1)
populationX[:,0]
```

```python
profitY = data['Profit'].values.reshape(-1,1)
profitY[:,0]
```

```python
X = np.matrix(populationX)
X.shape
```

```python
y = np.matrix(profitY)
y.shape
```

```python
#Para que espere una matriz (None, 1)
inputs = Input(shape=(1,))
#Linear para no especificar ningun activation
preds = Dense(1,activation='linear')(inputs)

model = Model(inputs, preds)
model.compile(optimizer='SGD' ,loss='mse')
model.fit(X,y, batch_size=70, epochs=1000)
```

```python
model.fit(X,y)
plt.scatter(X,y,color='black')
plt.plot(X ,model.predict(X), color='blue', linewidth=3)
```
