```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
df1 = pd.read_csv("data/tmdb_5000_credits.csv")
df2 = pd.read_csv("data/tmdb_5000_movies.csv")
df1.head()
```

```python
df2.head()
```

```python
df1.columns =["id","title","cast","crew"]
df2= df2.merge(df1, on="id")
```

```python
df2.head()
```

```python
C = df2['vote_average'].mean()
C
```

```python
m = df2['vote_count'].quantile(0.8)
m
'''I choose 0.8 to extract the movies that have more votes than the 80% of the total dataset'''
```

```python
q_movies = df2.copy().loc[df2['vote_count'] >=m]
q_movies.shape
```

```python
def weighted_rating(x, m=m, C=C):
    v = x['vote_count']
    R = x['vote_average']
    #Formula
    return (v/(v+m) * R) + (m/(m+v )* C)

q_movies['score'] = q_movies.apply(weighted_rating, axis=1)
```

```python
q_movies['title'] = q_movies['title_x']
q_movies.drop(['title_y'], axis=1)
q_movies.drop(['title_x'], axis=1)
```

```python
#Sort movies based on score calculated above
q_movies = q_movies.sort_values('score', ascending=False)

#Print the top 15 movies
q_movies[['title', 'vote_count', 'vote_average', 'score']].head(15)
```

```python
df2['title'] = df2['title_x']
df2.drop(['title_y'], axis=1)
df2.drop(['title_x'], axis=1)
df2.head(1)
```

```python
popularity = df2.sort_values("popularity", ascending=False)
plt.figure(figsize=(15,10))
plt.barh(popularity['title'].head(10), popularity['popularity'].head(10), color="red", align="center")
plt.xlabel("Popularity")
plt.title("Popular Movies")
```

```python
#Importing 
from sklearn.feature_extraction.text import TfidfVectorizer

#Define a TF-IDF Vectorizer Object. 
#Remove all english stop words such as 'the', 'a'
tfidf = TfidfVectorizer(stop_words="english")

#Replace NaN with an empty string
df2['overview'] = df2['overview'].fillna('')

#Construct the required TF-IDF matrix by fitting and transforming the data
tfidf_matrix = tfidf.fit_transform(df2['overview'])

tfidf_matrix.shape
```

```python
from sklearn.metrics.pairwise import linear_kernel

#Compute the cosine similarity matrix

cos_sim = linear_kernel(tfidf_matrix, tfidf_matrix)
```

```python
#Changing the index to the title of the movie for the recommendations
indices = pd.Series(df2.index, index=df2['title']).drop_duplicates()
```

```python
def get_recommendations(title, cosine_sim =cos_sim):
    # Get the index of the movie that matches the title
    idx = indices[title]
    
    # Get the pairs of the similarity scores of all movies with that movie
    similarity_scores = list(enumerate(cos_sim[idx]))
    
    # Sort the movies based on the similarity scores
    similarity_scores = sorted(similarity_scores, key=lambda x:x[1],reverse=True)
    # Get the scores of the 10 most similar movies
    similarity_scores = similarity_scores[1:6]
    
     # Get the movie indices
    movie_indices = [i[0] for i in similarity_scores]

    # Return the top 10 most similar movies
    return df2['title'].iloc[movie_indices]
    
```
